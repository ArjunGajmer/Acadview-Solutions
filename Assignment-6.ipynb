```python
##Import the libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
##Extract csv
df = pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Weather/master/weather.csv ")
```

```python
df.head()
```

```python
df.dtypes
```

```python
##Drop null values
df = df.dropna()
```

#### Examples of Univariate Plots
Plotted below are some univariate graphs. Feel free to add your own graphs.

```python
##Example 1
sns.distplot(df['Rainfall'], bins= 20)
```

```python
##Example 2
sns.kdeplot(df['Sunshine'], shade=True)
```

```python
##Example 3
sns.distplot(df['WindGustSpeed'], hist=False)
```

```python
##Example 4
sns.countplot(df['RainToday'])
```

```python
##Example 5
sns.boxplot(df['MaxTemp'])
```

#### Examples of Bivariate Plots

Plotted below are some bivariate graphs. Feel free to add your own graphs.

```python
##Example 1
sns.jointplot(x='Sunshine', y='Evaporation', data=df)
```

```python
##Example 2
sns.jointplot(x=df['Temp9am'], y=df['Humidity9am'], color='k', kind="hex")
```

```python
##Example 3
sns.jointplot(x=df['Cloud3pm'], y=df['WindSpeed3pm'], kind="kde")
```
